```python
import pandas as pd

file='data/leagues_NBA_2014_games_games.csv'

data = pd.read_csv(file,parse_dates=['Date'])
data.columns = ["Date",'Start Time',"Score Type", "Visitor Team", "VisitorPts", "Home Team", "HomePts", "OT?", "Notes"]

data.ix[:5]
```

```python
#数据清洗

#计算主场获胜
data['HomeWin'] = data['VisitorPts'] < data["HomePts"]

y_true = data['HomeWin'].values

data.ix[:5]
```

```python
from collections import  defaultdict 

#是否赢得上一场比赛
won_last = defaultdict(int)

for i, row in data.iterrows():
    home_team = row['Home Team']
    vistor_team = row['Visitor Team']
    row['HomeLastWin'] = won_last['home_team']
    row['VisitorLastWin'] = won_last[vistor_team]
    data.ix[i] = row 
    
    #更新主客场特征值
    won_last[home_team] = row['HomeWin']
    won_last[visitor_team] = not row['HomeWin']

```
